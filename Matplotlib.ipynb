```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Data for 3D plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='coolwarm')

# Adding titles
ax.set_title("3D Plot Example")
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]  # Squares
y2 = [25, 20, 15, 10, 5]  # Inverse trend

fig, ax1 = plt.subplots()

# Plot first dataset (y1)
ax1.plot(x, y1, 'g-')  # Green line
ax1.set_xlabel('X Axis')
ax1.set_ylabel('Squares', color='g')

# Create a second y-axis
ax2 = ax1.twinx()
ax2.plot(x, y2, 'b--')  # Blue dashed line
ax2.set_ylabel('Inverse', color='b')

plt.title("Dual Axis Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 4, 8, 16]
y2 = [0.5, 1.5, 2.5, 4.5, 7.5]

# Create an area plot
plt.fill_between(x, y1, y2, color="skyblue", alpha=0.4, label="Area 1")
plt.fill_between(x, y2, color="lightgreen", alpha=0.5, label="Area 2")

# Adding titles and labels
plt.title("Filled Area Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for heatmap
data = np.random.random((10, 10))  # Generate a 10x10 matrix of random values

# Create a heatmap
plt.imshow(data, cmap='viridis', interpolation='nearest')

# Add a colorbar to show the scale
plt.colorbar()

# Adding titles
plt.title("Heatmap Example")

# Show the plot
plt.show()
```
